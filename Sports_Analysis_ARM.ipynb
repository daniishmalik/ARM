```python
import matplotlib.pyplot as plt 
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US')

keywords = ['Football', 'Rugby', 'Tennis']

pytrends.build_payload(keywords, timeframe='today 3-m')

data = pytrends.interest_over_time()
data1 = pytrends.interest_by_region(resolution='COUNTRY')

print(data.head(10))
print(data1.head(10))
```

```python
# Average Monthly trend of Football, Rugby and Tennis on Google for last 3 months

df=data.reset_index()
df=df.iloc[:,0:4]
monthly_avg = df.resample(rule='ME', on='date')['Football','Rugby','Tennis'].mean()
print(monthly_avg)
monthly_avg.plot(kind = 'line', figsize = (5,4), title = 'Average Monthly trend of Football, Rugby and Tennis on Google')
```

```python
# Biweekly search trend comparison for Football, Rugby and Tennis on Google in last 3 months

plt.figure(figsize=(10,4))
plt.plot(data)
plt.legend(keywords, loc = 'upper right')
plt.suptitle('Biweekly search trend comparison for Football, Rugby and Tennis on Google in last 3 months')
plt.xlabel('Dates')
plt.ylabel('Trend Percentage')
plt.show()
```

```python
# Average Last 3 months trend using pie chart

avg_trend = round(data.mean(),2)
avg_trend.head(3).plot(kind = 'pie', figsize=(5,4), title ='Last 3 months trend rate comparison', xlabel=True)
```

```python
# Top 20 countries for Football search trend in last 3 months

top_twenty_football = data1.sort_values(by = ['Football'], ascending=False).head(20)

top_twenty_football = top_twenty_football.loc[:, "Football"]

top_twenty_football.plot(kind = 'bar', figsize=(20,4))
```
